```python
def h1(state):
    num_misplaced_tiles = 0
    for i in range(3):
        for j in range(3):
            if goal_state[i][j] != state[i][j]:
                num_misplaced_tiles = num_misplaced_tiles + 1
    
    return num_misplaced_tiles

def h2(state):
    manhattan_distance = 0
    goal_a = 0
    goal_b = 0
    for i in range(3):
        for j in range(3):
            if state[i][j] == goal_state[i][j]:
                manhattan_distance = manhattan_distance + 0
            elif state[i][j] != goal_state[i][j]:
                for a in range(3):    #search goal state
                    for b in range(3):
                        if goal_state[a][b] == state[i][j]:
                            goal_a = a
                            goal_b = b
                manhattan_distance = manhattan_distance + (abs(i - goal_a) + abs(j - goal_b))

    
    return manhattan_distance
```
